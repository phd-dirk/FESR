```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('../../FESR/configurations/2018/wCubeD6D8/fits.csv', delimiter=',', skip_header=2)
c6s = data[:,5]
c6Errs = data[:,6]
c8s = data[:,7]
c8Errs = data[:,8]
s0s = np.array([9, 8, 7, 6, 5, 4, 3])
plt.suptitle('(1-x)^3(1-3x) with D6 and D8 contributions')
plt.subplot(1, 2, 1)
plt.title('c6 with errors')
plt.errorbar(s0s, c6s, yerr=c6Errs, barsabove=True, fmt='.', marker='o')
plt.subplot(1, 2, 2)
plt.title('c8 with errors')
plt.errorbar(s0s, c8s, yerr=c8Errs, barsabove=True, fmt='.', marker='o')
plt.show()
bestWCubeD6D8Row = data[1,:]
```

```python
data = np.genfromtxt('../../FESR/configurations/2018/wCubeD6D8D10/fits.csv', delimiter=',', skip_header=2)
c6s = data[:,5]
c6Errs = data[:,6]
c8s = data[:,7]
c8Errs = data[:,8]
c10s = data[:,9]
c10Errs = data[:,10]
s0s = np.array([9, 8, 7, 6, 5, 4])
plt.suptitle('(1-x)^3(1-3x) with D6, D8 and D10 contributions')
plt.subplot(1, 3, 1)
plt.title('c6 with errors')
plt.errorbar(s0s, c6s, yerr=c6Errs, barsabove=True, fmt='.', marker='o')
plt.subplot(1, 3, 2)
plt.title('c8 with errors')
plt.errorbar(s0s, c8s, yerr=c8Errs, barsabove=True, fmt='.', marker='o')
plt.subplot(1, 3, 3)
plt.title('c10 with errors')
plt.errorbar(s0s, c10s, yerr=c10Errs, barsabove=True, fmt='.', marker='o')
plt.show()
bestWCubeD6D8D10Row = data[2,:]
```

```python
data = np.genfromtxt('../../FESR/configurations/2018/wCubeD6D8D10D12/fits.csv', delimiter=',', skip_header=2)
c6s = data[:,5]
c6Errs = data[:,6]
c8s = data[:,7]
c8Errs = data[:,8]
c10s = data[:,9]
c10Errs = data[:,10]
c12s = data[:,11]
c12Errs = data[:,12]
s0s = np.array([9, 8, 7, 6])
plt.suptitle('(1-x)^3(1-3x) with D6, D8, D10 and D12 contributions')
plt.subplot(2, 2, 1)
plt.title('c6 with errors')
plt.errorbar(s0s, c6s, yerr=c6Errs, barsabove=True, fmt='.', marker='o')
plt.subplot(2, 2, 2)
plt.title('c8 with errors')
plt.errorbar(s0s, c8s, yerr=c8Errs, barsabove=True, fmt='.', marker='o')
plt.subplot(2, 2, 3)
plt.title('c10 with errors')
plt.errorbar(s0s, c10s, yerr=c10Errs, barsabove=True, fmt='.', marker='o')
plt.subplot(2, 2, 4)
plt.title('c12 with errors')
plt.errorbar(s0s, c12s, yerr=c12Errs, barsabove=True, fmt='.', marker='o')
plt.show()
bestWCubeD6D8D10D12Row = data[1,:]
```

```python
data = np.genfromtxt('../../FESR/configurations/2018/wCubeD6D8Alpha/fits.csv', delimiter=',', skip_header=2)
alphas = data[:,1]
alphaErrs = data[:,2]
c6s = data[:,5]
c6Errs = data[:,6]
c8s = data[:,7]
c8Errs = data[:,8]
s0s = np.array([9, 8, 7, 6, 5])
plt.suptitle('(1-x)^3(1-3x) with D6, D8 contributions. Alpha is also fitted.') 
plt.subplot(1, 3, 1)
plt.title('alpha with errors')
plt.errorbar(s0s, alphas, yerr=alphaErrs, barsabove=True, fmt='.', marker='o')
plt.subplot(1, 3, 2)
plt.title('c6 with errors')
plt.errorbar(s0s, c6s, yerr=c6Errs, barsabove=True, fmt='.', marker='o')
plt.subplot(1, 3, 3)
plt.title('c8 with errors')
plt.errorbar(s0s, c8s, yerr=c8Errs, barsabove=True, fmt='.', marker='o')
plt.show()
bestWCubeD6D8AlphaRow = data[2,:]
```

```python
bestRows = np.matrix([bestWCubeD6D8Row, bestWCubeD6D8D10Row, bestWCubeD6D8D10D12Row, bestWCubeD6D8Alpha])
c6s = bestRows[:,5]
c6Errs = bestRows[:,6]
c8s = bestRows[:,7]
c8Errs = bestRows[:,8]
x = np.array(['6,8', '6,8,10', '6,8,10,12', '6,8,al'])
plt.subplot(1, 2, 1)
plt.title('c6s with errors')
plt.errorbar(x, c6s, yerr=c6Errs, barsabove=True, fmt='.', marker='o')
plt.subplot(1, 2, 2)
plt.title('c8s with errors')
plt.errorbar(x, c8s, yerr=c8Errs, barsabove=True, fmt='.', marker='o')
plt.show()
```
